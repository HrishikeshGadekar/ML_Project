#### 4) Data Cleaning

Now we will move on to the next step of cleaning the data by checking for any missing or infeasible data. This will make the Dataset make suitable for applying Machine Learning algorithms

As we can infer from the Distribtution plot of 'Experience' column, some of the values are negative, which is not possible since the range of the variable 'Experience' is [0, inf]. We will set the negative values to the mmedian of the remaining values in that column. The reason for choosing Median and not Mean is that Mean is sensitive to outliers in the Dataset while the Median is not.

```python
df[df['Experience'] < 0]['Experience'].count()
```

```python
lmdvalue = int(df['Experience'].median())
dlf = df[df['Experience'] < 0]['Experience']
for i in range(len(df)):
    if df['Experience'].iloc[i] < 0:
        df['Experience'].iloc[i] = lmdvalue
df['Experience']
```
